# Using Xpublish Endpoints

Xpublish supports serving environmental data through a number of plugins. To view the current capabilities of an xpublish endpoint, you can browse to the /docs page, e.g. http://nextgen-dev.ioos.us/xreds/docs

## OpenDAP Endpoints

[xpublish-opendap](https://github.com/xpublish-community/xpublish-opendap) is a plugin that defines the methods for the OpenDAP protocol.

Examples:
- http://nextgen-dev.ioos.us/xreds/datasets/dbofs/opendap.dds
- http://nextgen-dev.ioos.us/xreds/datasets/dbofs/opendap.das
- http://nextgen-dev.ioos.us/xreds/datasets/dbofs/opendap.dods


```python
import xarray as xr

url = 'http://nextgen-dev.ioos.us/xreds/datasets/dbofs/opendap'

ds = xr.open_dataset(url)
ds
```
